```python
Employee['Full Name'] = Employee['FirstName'] + ' ' + Employee['LastName']
Employee.drop(['FirstName', 'LastName','Phone','Email','PostalCode','Fax'], axis=1, inplace=True)
```

```python
# Convert 'BirthDate' and 'HireDate' to datetime
Employee['BirthDate'] = pd.to_datetime(Employee['BirthDate'])
Employee['HireDate'] = pd.to_datetime(Employee['HireDate'])

# Calculate age
today = pd.Timestamp('now')
Employee['Age'] = ((today - Employee['BirthDate']) / pd.Timedelta(days=365.25)).astype(int)

# Calculate years of work
Employee['YearsOfWork'] = ((today - Employee['HireDate']) / pd.Timedelta(days=365.25)).astype(int)

# Display the updated DataFrame
Employee
```

```python
Employee = Employee.fillna(0)
```
